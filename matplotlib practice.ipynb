```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis,y_axis)
```

```python
# create a plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# create a plot with ax.plt() using alternative way
fig = plt.figure()
ax=fig.add_subplot()
```

```python
#create the plot
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# create the plot and add a legend
plt.plot(x_axis,y_axis,marker="*", color="blue", linewidth=2, label='Boston')
# create lABEL for x axis and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Create a title.
plt.title("Pyber fare per month")
#add the legend
plt.legend()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax=fig.add_subplot()
ax.plot(x_axis,y_axis,'g-', marker="D",lw=2,label="boston")
ax.set_title("fare company")
ax.set_xlabel('fate')
ax.set_ylabel('date')
ax.legend()
ax.color="blue"
ax.grid()
```

```python
# create the plot
plt.bar(x_axis,y_axis,color = 'green',label ='boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Create the plot with ax.plt() object oriented method
fig, ax=plt.subplots()
ax.bar(x_axis,y_axis, color="black", label= "New York")
```

```python
fig, ax=plt.subplots()
ax.barh(x_axis,y_axis, color="cyan", label="chicago")
ax.set_xlabel("fare($)")
ax.set_ylabel("date")
ax.set_title("pyber fare by month")
ax.set_xticks(np.arange(0,50,step=5.0))
ax.invert_yaxis()
ax.legend()
```

```python
#Add Minor Ticks
from matplotlib.ticker import MultipleLocator
#Increase the size of the plot figure.
fig,ax=plt.subplots(figsize=(8,8))
ax.barh(x_axis,y_axis,label='chicago')
ax.set_xticks(np.arange(0,50,step=5.0))
ax.set_xlabel("fare($)")
ax.set_ylabel("date")
ax.set_title("pyber fare by month")
ax.invert_yaxis()
ax.legend()
# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
```

```python
#Scatter Plot Using the MATLAB Method - plt.plot()
plt.plot(y_axis,x_axis,'o',color='red',label= 'chicago')
plt.xlabel("fare")
plt.ylabel("date")
plt.title("pyber fare by month")
plt.xlim(0,45)
plt.gca().invert_yaxis()
plt.legend()
```

```python
#Scatter Plot Using the MATLAB Method - plt.scatter()
plt.scatter(x_axis, y_axis)
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#  list comprehension to replace many for and while loops
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis,color ='blue',marker ='D', s=y_axis, aplha=0.2)
```

```python
plt.pie(y_axis, labels=x_axis)
```

```python
plt.subplots(figsize=(8,8))
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis,explode=explode_values,labels=x_axis,colors=colors,autopct='%.1f%%')

plt.show()
```

```python
fig, ax =plt.subplots(figsize=(8,8))
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
ax.pie(y_axis, labels=x_axis,autopct='%.1f%%',explode=explode_values,shadow='True',startangle=90,counterclock=False)
plt.show()
```
